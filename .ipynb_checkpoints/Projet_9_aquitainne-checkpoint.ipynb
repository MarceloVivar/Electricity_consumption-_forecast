## Prédisez la demande en électricité

## Mise en situation

Vous êtes employé chez Enercoop, société coopérative qui s'est développée grâce à la libéralisation du marché de l’électricité en France. Elle est spécialisée dans les énergies renouvelables.

La plupart de ces énergies renouvelables est cependant intermittente, il est donc difficile de prévoir les capacités de production d'électricité. De plus, la demande en électricité des utilisateurs varie au cours du temps, et dépend de paramètres comme la météo (température, luminosité, etc.) Tout le challenge est de mettre en adéquation l'offre et la demande !

## Les données


Vous téléchargerez les données de consommation d'électricité mensuelles à partir de cette page : http://www.rte-france.com/fr/eco2mix/eco2mix-telechargement

Les données météo que vous utiliserez pour corriger les données de l'effet température sont présentes ici : https://cegibat.grdf.fr/simulateur/calcul-dju

## Votre mission

Vous vous concentrerez uniquement sur la prédiction de la demande en électricité.

1. Corrigez les données de consommation mensuelles de l'effet température (dues au chauffage électrique) en utilisant une régression linéaire.

### 3.Effectuez une prévision de la consommation (corrigée de l'effet température) sur un an, en utilisant la méthode de Holt Winters (lissage exponentiel) puis la méthode SARIMA sur la série temporelle.

### La méthode de Holt Winters (lissage exponentiel) 

```python
y=np.asarray(df_corr["corrigee"])


hw = ExponentialSmoothing(y, seasonal_periods=12, trend='add', seasonal='add').fit()
hw_pred = hw.forecast(12)

```

```python
hw.params
```

```python
plt.figure(figsize=(14,6))
plt.plot(df_corr['Mois'],df_corr['corrigee'], label='Consommation Corrigee')
plt.plot(pd.date_range(df_corr.Mois[df_corr.shape[0]-12],
                       periods=12, freq='M'), hw_pred, label='Predit')
plt.legend()
plt.show()
plt.savefig('projet_9_images/Conso_corrigée_predit.png')
```

```python
hw_pred
```

```python
# plot de la prediction
x_tronc = df_corr.iloc[-12:,:]
x_tronc['prediction']=hw_pred.copy()
```

```python
plt.figure(figsize=(14,6))
plt.plot(x_tronc['Mois'],x_tronc['prediction'], label='prediction')
plt.plot(x_tronc['Mois'],x_tronc['corrigee'], label='true values')
plt.legend()
plt.savefig('projet_9_images/Conso_corrigée_predit_zoom.png')
```

```python
stat, p = scipy.stats.kstest(hw.resid,'norm')
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpretion
alpha = 0.05
if p > alpha:
	print('L’échantillon semble gaussien (ne pas rejeter H0)')
else:
	print('L’échantillon ne semble pas gaussien (rejeter H0)')
```

### Modèle 1

```python
df_corr.index = df_corr['Mois']
```

```python
y= df_corr["corrigee"].values
```

```python
from statsmodels.tsa.statespace.sarimax import *
from statsmodels.stats.diagnostic import acorr_ljungbox

model1 = SARIMAX(y, order=(1,1,1), seasonal_order=(1,1,1,12))
results1 = model1.fit()
print(results1.summary())

print('Retard : p-value')
for elt in [6, 12, 18, 24, 30, 36]:
    print('{} : {}'.format(elt, acorr_ljungbox(results1.resid, lags=elt)[1].mean()))
```

```python
results1.plot_diagnostics()
plt.show()
```

```python
stat, p = scipy.stats.kstest(results1.resid,'norm')
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpretion
alpha = 0.05
if p > alpha:
	print('L’échantillon semble gaussien (ne pas rejeter H0)')
else:
	print('L’échantillon ne semble pas gaussien (rejeter H0)')
```

```python
pred_model1 = results1.get_forecast(12)
pred_1 = pred_model1.predicted_mean
pred_l = [elt[0] for elt in pred_model1.conf_int(alpha=0.05)]
pred_u = [elt[1] for elt in pred_model1 .conf_int(alpha=0.05)]

plt.figure(figsize=(14,6))
plt.plot(df_corr["corrigee"], label='Consommation corrigee')
plt.plot(pd.date_range(df_corr.index[len(df_corr)-12], periods=12, freq='M'), pred_1, color='r', label='Predit')
plt.plot(pd.date_range(df_corr.index[len(df_corr)-12], periods=12, freq='M'), pred_l, color='g', linestyle='--')
plt.plot(pd.date_range(df_corr.index[len(df_corr)-12], periods=12, freq='M'), pred_u, color='g', linestyle='--')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14,6))
plt.plot(df_corr["corrigee"]['2018':], label='Consommation corrigee')
plt.plot(pd.date_range(df_corr.index[len(df_corr)-12], periods=12, freq='M'), pred_1, color='r', label='Predit')
plt.plot(pd.date_range(df_corr.index[len(df_corr)-12], periods=12, freq='M'), pred_l, color='g', linestyle='--')
plt.plot(pd.date_range(df_corr.index[len(df_corr)-12], periods=12, freq='M'), pred_u, color='g', linestyle='--')
plt.legend()
plt.show()
```

```python
def tsplot(y, lags=None, figsize=(10, 8), style='bmh'):
    if not isinstance(y, pd.Series):
        y = pd.Series(y)
    with plt.style.context(style):    
        fig = plt.figure(figsize=figsize)
        #mpl.rcParams['font.family'] = 'Ubuntu Mono'
        layout = (3, 2)
        ts_ax = plt.subplot2grid(layout, (0, 0), colspan=2)
        acf_ax = plt.subplot2grid(layout, (1, 0))
        pacf_ax = plt.subplot2grid(layout, (1, 1))
        qq_ax = plt.subplot2grid(layout, (2, 0))
        pp_ax = plt.subplot2grid(layout, (2, 1))
        
        y.plot(ax=ts_ax)
        ts_ax.set_title('Time Series Analysis Plots')
        smt.graphics.plot_acf(y, lags=lags, ax=acf_ax, alpha=0.5)
        smt.graphics.plot_pacf(y, lags=lags, ax=pacf_ax, alpha=0.5)
        sm.qqplot(y, line='s', ax=qq_ax)
        qq_ax.set_title('QQ Plot')        
        scs.probplot(y, sparams=(y.mean(), y.std()), plot=pp_ax)

        plt.tight_layout()
    return 
```

```python
tsplot(y, lags=12)
```

```python

am = arch_model(y)
res = am.fit(update_freq=5)
p(res.summary())
```



* Iterate through combinations of ARIMA(p, d, q) models to best fit our time series.

* Pick the GARCH model orders according to the ARIMA model with lowest AIC.

* Fit the GARCH(p, q) model to our time series.

* Examine the model residuals and squared residuals for autocorrelation

```python


def _get_best_model(TS):
    best_aic = np.inf 
    best_order = None
    best_mdl = None

    pq_rng = range(3) # [0,1,2]
    d_rng = range(2) # [0,1]
    for i in pq_rng:
        for d in d_rng:
            for j in pq_rng:
                try:
                    tmp_mdl = smt.ARIMA(TS, order=(i,d,j)).fit(
                        method='mle', trend='nc'
                    )
                    tmp_aic = tmp_mdl.aic
                    if tmp_aic < best_aic:
                        best_aic = tmp_aic
                        best_order = (i, d, j)
                        best_mdl = tmp_mdl
                except: continue
    p('aic: {:6.5f} | order: {}'.format(best_aic, best_order))                    
    return best_aic, best_order, best_mdl

# Notice I've selected a specific time period to run this analysis

res_tup = _get_best_model(y)

```

```python
# Now we can fit the arch model using the best fit arima model parameters

p_ = 1
o_ = 1
q_ = 2

# Using student T distribution usually provides better fit
am = arch_model(y, p=p_, o=o_, q=q_, dist='StudentsT')
res = am.fit(update_freq=5, disp='off')
p(res.summary())
```

```python
stat, p = scipy.stats.jarque_bera(res.resid)
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpretion
alpha = 0.05
if p > alpha:
    print('L’échantillon semble gaussien (ne pas rejeter H0)')
else:
    print('L’échantillon ne semble pas gaussien (rejeter H0)')
```

```python
fig = res.plot(annualize='D')
```

```python
pred_model3 = res.forecast(horizon=1, start=df_corr['corrigee_saison']['2018':], method='simulation')
pred_3 = pred_model3.predicted_mean
pred_l = [elt[0] for elt in pred_model3.conf_int(alpha=0.05)]
pred_u = [elt[1] for elt in pred_model3 .conf_int(alpha=0.05)]

plt.figure(figsize=(14,6))
plt.plot(df_corr["corrigee"], label='Consommation corrigee')
plt.plot(pd.date_range(df_corr.index[len(df_corr)-12], periods=12, freq='M'), pred_3, color='r', label='Predit')
plt.plot(pd.date_range(df_corr.index[len(df_corr)-12], periods=12, freq='M'), pred_l, color='g', linestyle='--')
plt.plot(pd.date_range(df_corr.index[len(df_corr)-12], periods=12, freq='M'), pred_u, color='g', linestyle='--')
plt.legend()
plt.show()
```
